# Image Library

```python
# pip install Pillow
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
img= np.array(Image.open("igi.png"))
type(img)
```

```python
# print("Format: ",img.format,"\nSize: ", img.size,"\nMode: ", img.mode)
plt.imshow(img)
```

```python
img.shape
```

```python
# getting r depth of image
r = img[:, :, 0]
print(r)
plt.imshow(r)
```

```python
# getting green depth of image
g = img[:, :, 1]
plt.imshow(g)
```

```python
# getting blue depth of image matrix 
# and plotting it

b = img[:, :, 2]
plt.imshow(b)
```

```python
img = Image.open("igi.png")
img.thumbnail((64, 64))
implot = plt.imshow(img)
implot
```

```python
img = Image.open("igi.png")
img.thumbnail((64, 64), resample = Image.Resampling.BICUBIC )
implot = plt.imshow(img)
implot
```

```python
imgplot = plt.imshow(img, interpolation="bilinear")
```

```python
imgplot = plt.imshow(img, interpolation="bicubic")
```
