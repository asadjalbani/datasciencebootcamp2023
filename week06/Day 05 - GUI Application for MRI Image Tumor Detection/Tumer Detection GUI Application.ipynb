```python
import os
import tkinter as tk
from tkinter import filedialog
from tkinter import messagebox
from PIL import Image, ImageTk
import numpy as np
import tensorflow as tf
```

```python
# Load a saved model from a same file BrainTumor.h5
model = tf.keras.models.load_model ('./BrainTumor.h5')
```

```python
# Create the main window
root = tk.Tk()
root.title("Tumor Detection GUI")
```

```python
# Function to Classify the image
def classify_image():
    # Define the file types as a list of tuples
    filetypes = [
        ("Image Files", "*.png;*.jpg;*.jpeg"),  # Description and file patterns separated by a semicolon
    ]
    file_path = filedialog.askopenfilename(filetypes=filetypes)
    if file_path:
        image = Image.open(file_path).resize((128, 128))
        image_array = np.array(image)
        prediction = model.predict(np.array([image_array]))
        predicted_tumor = np.max(prediction)

         if prediction==0:
            prediction='Gloma Tumor'
        elif prediction==1:
            prediction='Meningioma Tumor'
        elif prediction==2:
            prediction='No Tumor'
        elif prediction==3:
            prediction='Petituray Tumor'

        image_label.config(image=ImageTk.PhotoImage(image))
        result_label.config(text=f"Predicted Tumor : {predicted_tumor}")
```

```python

open_button = tk.Button(root, text="Open Image", command=classify_image)
image_label = tk.Label(root)
result_label = tk.Label(root)
```

```python
open_button.pack(pady=10)
image_label.pack()
result_label.pack()
```

```python
# Run the Tkinter main loop
root.mainloop()
```
