```python
# Import necessary libraries
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# Sample data (documents and their corresponding labels)
documents = [
    'This is the first document.',
    'This document is the second document.',
    'And this is the third one.',
    'Is this the first document?',
    'The sky is blue.',
    'I love programming.',
    'Programming is fun.',
    'Machine learning is a subfield of AI.',
    'I enjoy learning about NLP.',
    'NLP models can process text data.'
]

labels = [0, 1, 0, 1, 0, 1, 1, 0, 1, 0]  # Example labels (0 for class 1, 1 for class 2)
```

```python
# Step 1: Convert text data to BoW representation
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(documents)
```

```python
# Step 2: Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, random_state=42)
```

```python
# Step 3: Train a simple Naive Bayes classifier
classifier = MultinomialNB()
classifier.fit(X_train, y_train)
```

```python
# Step 4: Make predictions on the test set
y_pred = classifier.predict(X_test)
```

```python

# Step 5: Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')
```
