```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.model_selection import cross_val_score
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_openml
```

```python
X, y = fetch_openml('mnist_784', version = 1, return_X_y=True, as_frame=False)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
print(X.shape)
print(y.shape)
```

```python
classifier = DecisionTreeClassifier(max_depth=8)
```

```python
my_model = classifier.fit(X_train, y_train)
y_pred = my_model.predict(X_test)
my_model_score = accuracy_score(y_test, y_pred)
```

```python
k_fold = 5
cross_val_scores = cross_val_score(classifier, X, y, cv=k_fold)
```

```python
# printing the result

print('Cross-Validation Scores:', cross_val_scores)
print('Mean Cross_Validation Score:', np.mean(cross_val_scores))
print('My Model Score:', my_model_score)
```
