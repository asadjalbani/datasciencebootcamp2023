# Testing the saved Multinomial Model on own generated data

```python
# import the libraries
import joblib as jl
from PIL import Image, ImageFilter, ImageEnhance
import numpy as np
import matplotlib.pyplot as plt
```

```python

# load saved model 
model = jl.load('knn')

# load the image 
img = Image.open('2.jpeg')
plt.imshow(img)
```

```python
# convert image to grayimage
gray_image = img.convert('L')
plt.imshow(gray_image, cmap='gray')
```

```python
img_pixel = np.array(gray_image)

# negative the image
img_neg = 255 - img_pixel

# create new image form neg image
img_n = Image.fromarray(img_neg)

# show image
plt.imshow(img_n, cmap='gray')
```

```python
# enhance the edge of image
img_enhance = img_n.filter(ImageFilter.EDGE_ENHANCE)

gray_image = img_enhance 
plt.imshow(gray_image)
```

```python
# resize the image 8 by 8
img_reshape = np.array(gray_image.resize((8,8)))
img_fl = img_reshape.flatten()
plt.imshow(img_reshape, cmap='gray')
```

```python
# shape and size of the image
print(img_fl.shape)
print(img_fl.size)
```

```python
test = img_fl.reshape(1,-1)
```

# Testing the model

```python
model.predict(test)
```
